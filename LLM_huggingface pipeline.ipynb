```python
텍스트생성 (text generation)
```

```python
from transformers import pipeline

# 텍스트 생성 파이프라인 설정
generator = pipeline('text-generation', model='gpt2')

# 텍스트 입력 결과 생성
result = generator("Once upon a time")

print(result)
```

감성 분석

```python
from transformers import pipeline

# 텍스트 생성 파이프라인 설정
generator = pipeline('sentiment-analysis')

# 텍스트 입력 결과 생성
result = generator("I love this product")

print(result)
```

질문-답변

```python
from transformers import pipeline

# 텍스트 생성 파이프라인 설정
qa_pipeline  = pipeline('question-answering')

# 텍스트 입력 결과 생성
result = qa_pipeline({
    'question': 'What is the capital of France?',
    'context': 'France is a country in Europe. Its capital is Paris'
})

print(result)
```

번역 실행

```python
from transformers import pipeline

# 텍스트 생성 파이프라인 설정
translator  = pipeline('translation_en_to_fr')

# 텍스트 입력 결과 생성
text = "Hugging Face is creating amazing tools for developers and researchers"

translation = translator(text)
print(translation)
```

```python
from transformers import pipeline

# 텍스트 생성 파이프라인 설정
summarizer  = pipeline('summarization')

# 텍스트 입력 결과 생성
text = """
The Transformer architecture has been revolutionary in natural language processing.
It has allowed for the creation of models that can understand and generate human language with incredible accuracy.
These models, such as BERT and GPT, have been applied to a wide variety of tasks including text classification,
machine translation, and even generating new creative text.
"""

summary = summarizer(text, max_length=20, min_length=10, do_sample = False)
print(summary)
```
