```python
from huggingface_hub import login
from transformers import AutoModelForCausalLM, AutoTokenizer

#모델과 토크나이저 설정
#model_name = "openai-community/gpt2"  #Hugging Face에서 사용할 모델이름 (https://huggingface.co/openai-community/gpt2 에서 복사한 이름)
#model_name = "skt/kogpt2-base-v2"
model_name = "meta-llama/Llama-3.2-1B"   #허깅페이스 모델사용신청서 허가받은후 사용가능함

# 허깅페이스 메일
login("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") #로그인 토큰 -추가코드 필요한지 확인해야함

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

```

Step3. 챗봇 함수 작성

```python
import torch

def generate_response(prompt) :
    #사용자 입력을 토큰화
    inputs = tokenizer(prompt, return_tensors="pt")

    #모델로 텍스트생성
    outputs = model.generate(
        inputs["input_ids"],
        max_length=100,             #최대 생성 길이 설정
        num_return_sequences=1,     #한 개의 응답만 생성
        no_repeat_ngram_size=2,     #반복방지
        top_p=0.9,                  #상위 확률 샘플링
        do_sample=True              #샘플링활성화
    )

    #토큰을 텍스트로 디코딩
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

```

Step4. 대화테스트

```python
print("챗봇과 대화를 시작해보세요 종료라고 입력하면 종료됩니다")

while True:
    user_input = input("You :")
    if user_input.lower() == "종료":
        break

    response = generate_response(user_input)
    print("챗봇 :", response)

```
